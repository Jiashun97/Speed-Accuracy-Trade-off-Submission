```python
datasets_list = ["Dutilh2019", "Dutilh2013", "Evans2017", "EJ2008", "Rae2014", "Lerche2017"]
models_list = ["DDMOUM", "DDM", "OUM"]
df_all = pd.DataFrame(columns = ["Parameter", "Cohen's d", "Dataset", "Model"])

for model in models_list:
    for dataset in datasets_list:
        df = pd.read_csv(model + "result" + dataset + ".csv")
        df_dif = get_difference(dataset, df, model)
        df_cohen_d = get_cohen_d(df_dif,model)
        df_all = df_all.append(df_cohen_d)
```
