```python
def plot_01quantile_all_datasets(Condition, title, ax = None):
    if ax is None:
        ax = plt.gca()
    
    datasets = ["Dutilh 2013", "Dutilh 2019", "Evans 2017", "Lerche 2017", "Rae 2014", "Wagenmakers 2008"]

    all_dataset = pd.DataFrame()
    for dataset in datasets:
        quantiles_data = pd.read_csv("1quantiles_data_"+ dataset.replace(" ", ""))
        quantiles_DDM  = pd.read_csv("1quantiles_DDM_" + dataset.replace(" ", ""))
        quantiles_OUM  = pd.read_csv("1quantiles_OUM_" + dataset.replace(" ", ""))
        single_dataset = pd.DataFrame({"Data":quantiles_data[Condition], "DDM":quantiles_DDM[Condition], "OUM":quantiles_OUM[Condition]})
        single_dataset["Dataset"] = dataset

        all_dataset = pd.concat((all_dataset,single_dataset))

    all_dataset_melted = pd.melt(all_dataset, value_vars=['Data', 'DDM','OUM'],
                                 var_name = "Model", value_name = "0.1 RT Quantile", id_vars = ["Dataset"])
    #fig, ax = plt.subplots()
    sns.barplot(ax=ax, data=all_dataset_melted, x="Dataset", y="0.1 RT Quantile", hue="Model", legend = False)#, hue_order=["Speed","Accuracy"])
    #ax.set_ylim(0, 2.5)
    ax.set_title(title)
    ax.set_xlabel("")
    return ax
```

```python
%run SAT_help_functions.ipynb
fig, axes = plt.subplots(2,2,figsize=(20, 8))
plot_01quantile_all_datasets(Condition = "cor_s", title = "Correct Choices (Speed Condition)", ax=axes[0,0])
plot_01quantile_all_datasets(Condition = "err_s", title = "Incorrect Choices (Speed Condition)", ax=axes[0,1])
plot_01quantile_all_datasets(Condition = "cor_a", title = "Correct Choices (Accuracy Condition)", ax=axes[1,0])
plot_01quantile_all_datasets(Condition = "err_a", title = "Incorrect Choices (Accuracy Condition)", ax=axes[1,1])
```
