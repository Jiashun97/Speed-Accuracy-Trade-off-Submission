```python
datasets = ["Dutilh2019", "Rae2014", "Lerche2017", "Dutilh2013", "Evans2017", "Wagenmakers2008"] 

f, ax = plt.subplots(2, 3,figsize=(12, 8))#, sharey=True)
for i in range(3):

    df_DDM = pd.read_csv("DDM2v2a2c2terresult"+datasets[i]+".csv")
    df_OUM = pd.read_csv("OUMfullresult"+datasets[i]+".csv")
    dif_DDM = df_DDM["Collapse Speed"] - df_DDM["Collapse Accuracy"]
    dif_OUM = df_OUM["Self-excitation Speed"] - df_OUM["Self-excitation Accuracy"]
    ax[0,i].scatter(dif_DDM, dif_OUM, s = 20, facecolors='none', edgecolors='k')
    ax[0,i].set_title(datasets[i])
    
for i in range(3):
    df_DDM = pd.read_csv("DDM2v2a2c2terresult"+datasets[i+3]+".csv")
    df_OUM = pd.read_csv("OUMfullresult"+datasets[i+3]+".csv")
    dif_DDM = df_DDM["Collapse Speed"] - df_DDM["Collapse Accuracy"]
    dif_OUM = df_OUM["Self-excitation Speed"] - df_OUM["Self-excitation Accuracy"]
    ax[1,i].scatter(dif_DDM, dif_OUM, s = 20, facecolors='none', edgecolors='k')
    ax[1,i].set_title(datasets[i+3])
    
    ax[1,i].set_xlabel("Difference in OUM Self-excitation",fontsize=10)
    ax[0,0].set_ylabel("Difference in DDM Collapse Rate",fontsize=10)
    ax[1,0].set_ylabel("Difference in DDM Collapse Rate",fontsize=10)

#f[0,1].figure(figsize = (4,4))
```
